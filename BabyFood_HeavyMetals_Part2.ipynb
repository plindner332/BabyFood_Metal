## *1.  Rice free snacks have 93% less Toxic heavy metals than Puff snacks(rice).*
##### To compare these two first we will find the average toxic heavy metal value for:
Puff Snacks:  
- 'Snack-ricepuffs',
- 'Snack-puffs,containsrice'

and Rice free snacks:  
- 'Snack-puffs,non-rice',
- 'Snack-other'

```python
df.loc[(df['Food type'] == 'Snack-ricepuffs') | (df['Food type'] == 'Snack-puffs,containsrice'), 'Part1'] = 'Puff snacks(rice)' 
df.loc[(df['Food type'] == 'Snack-puffs,non-rice') | (df['Food type'] == 'Snack-other'), 'Part1'] = 'Rice free snacks' 
df.head()
```

```python
Type = df.groupby('Food type').mean() 
Type['Count']= df.groupby(['Food type']).Brand.count()
Type.head()
```

```python
array = ['Snack-ricepuffs', 'Snack-puffs,containsrice', 'Snack-puffs,non-rice', 'Snack-other']
Type.loc[Type.index.isin(array)]
```

```python
Type1 = df.groupby('Part1').mean() 
Type1['Count']= df.groupby(['Part1']).Brand.count()
Type1.head()
```

```python
Type1 = Type1.T
Type1.columns
```

```python
Type1['Percent drop: Rice --> Rice Free']= ((1 - Type1['Rice free snacks']/Type1['Puff snacks(rice)'])*100).round(0)
Type1.head(6)
```

 #### The finding I showed was a 92% drop of Arsenic and Mercury which seems reasonable to the claims from the article stating that Rice free had 93% less Toxic heavy metals.
 
From page 13: "Note: For each pair of foods shown, concentrations shown and the comparative term “less toxic metals” are based on the average of the sum of four metals (inorganic arsenic, lead, cadmium and mercury) for the available samples of each food, unless noted otherwise. Averages were computed using data from the current study combined with data from FDA’s market basket study (the Total Diet Study, FDA 2014-2017)."

##### This explains why it does match exactly the point or quanity cited in the study.  This makes it impossible to do check 2 or 4.

## 3.  Other infant cereals have 84% less Toxic Heavy metals than infant rice cereals.
##### To compare these two first we will find the average toxic heavy metal value for:
Infant Rice Cereals:  
- 'Cereal-rice'

Infant Cereals:  
- 'Cereal-mixedandmulti-grain',
- 'Cereal-oatmeal',
- 'Cereal-othersingle-grain'

```python
df.drop(columns=['Part1'])
```

```python
df.loc[(df['Food type'] == 'Cereal-rice'), 'Part3'] = 'Infant Rice Cereals' 
df.loc[(df['Food type'] == 'Cereal-mixedandmulti-grain') | 
       (df['Food type'] == 'Cereal-oatmeal') | (df['Food type'] == 'Cereal-othersingle-grain'), 'Part3'] = 'Infant Cereals' 
df.head()

Type = df.groupby('Food type').mean() 
Type['Count']= df.groupby(['Food type']).Brand.count()

array = ['Cereal-rice', 'Cereal-mixedandmulti-grain', 'Cereal-oatmeal', 'Cereal-othersingle-grain']
Type = Type.loc[Type.index.isin(array)]
Type.head()
```

```python
Type3 = df.groupby('Part3').mean() 
Type3['Count']= df.groupby(['Part3']).Brand.count()
Type3.head()
```

```python
Type3 = Type3.T
Type3.columns

Type3['Percent drop: Rice --> Rice Free']= ((1 - Type3['Infant Cereals']/Type3['Infant Rice Cereals'])*100).round(0)
Type3.head(6)
```

#### Again this is only looking at the data looking the cereals with out rice have a drop for all metal categories, except cadmium.  Following the summation of the 4 toxic metals (exclusing Arsenic (total, ppb)), the drop is 48% toxic metals.  The discrepancy from the study is due to the additional data source not included.

**Sweet Potatoes:**

```python
df.drop(columns=['Part3'])
```

```python
df.loc[(df['Food type'] == 'Veggie-single-carrot'), 'Part5'] = 'Carrots' 
df.loc[(df['Food type'] == 'Veggie-single-sweetpotato'), 'Part5'] = 'Sweet Potatoes' 
df.loc[(df['Food type'] == 'Fruitandveggie-mixed') | (df['Food type'] == 'Veggie-mixed'), 'Part5'] = 'Fruit & Veggie variety' 
#df.loc[(df['Food type'] == 'Veggie-single-other') | (df['Food type'] == 'Fruit-single-apple') | 
#       (df['Food type'] == 'Fruit-single-banana') | (df['Food type'] == 'Fruit-single-other') | 
#       (df['Food type'] == 'Fruit-single-pear') | (df['Food type'] == 'Fruit-single-prune'), 'Part5'] = 'Fruit & Veggie single' 
df.head()

Type = df.groupby('Food type').mean() 
Type['Count']= df.groupby(['Food type']).Brand.count()

array = ['Veggie-single-carrot', 'Veggie-single-sweetpotato', 'Fruitandveggie-mixed', 'Veggie-mixed']

Type = Type.loc[Type.index.isin(array)]
Type.head()
```

```python
Type5 = df.groupby('Part5').mean() 
Type5['Count']= df.groupby(['Part5']).Brand.count()
Type5.head()
```

```python
Type5 = Type5.T
Type5.columns

Type5['Carrots --> Variety']= ((1 - Type5['Fruit & Veggie variety']/Type5['Carrots'])*100).round(0)
Type5['Sweet Potatoes --> Variety']= ((1 - Type5['Fruit & Veggie variety']/Type5['Sweet Potatoes'])*100).round(0)
Type5.head(6)
```

#### From this data it doesn't show the conclusion clearly.  
Carrots compared to Variety shows a decrease in lead, Cadmium and Mercury but an increase in Arsenic (total, ppb).  

Sweet Potatoes compared to Variety shows a decrease in Lead and Arsenic (total, ppb) but an increase in Cadmium and Mercury.
